```python
from perceptron import Perceptron
import random
import numpy as np
from sklearn.datasets import load_iris
```

## Perceptrons and Train functions
Eerst maken we de perceptrons aan met willekeurig gekozen weights. De AND perceptron heeft een learning rate van 0.1, maar de XOR niet. Dit komt omdat een beter resultaat (lagere MSE) bereikt wordt met een 0.01 learning rate bij de XOR. Als de learning rate niet 0.01 is gaat de MSE omhoog naar 0.5.

```python
p1 = Perceptron("AND", [4, 9], 0, 0.1)
p2 = Perceptron("XOR", [3, 4], 0, 0.01)
```

Hier maken we een lijst met de mogelijke combinaties van inputs voor de perceptrons. Ook hebben we hier een kleine functie die het trainproces start en de informatie van de perceptrons print (weights, bias, learning_rate, etc.). <b>Zie perceptron.py voor meer informatie over de werking.</b>

```python
pos_2x = [[1, 1], [1, 0], [0, 1], [0, 0]]
```

```python
def train_perceptron(p, inputs, expected, num_of_epochs):
    print(f"{p}\n")
    p.train(inputs, expected, num_of_epochs)
    print(f"\n{p}")
```

### Training Perceptrons (AND & XOR)
Hier trainen we en testen we de eerste twee cases, de AND en XOR. We geven dus de perceptron, inputs, verwachte waardes en hoeveelheid epochs mee.
We kunnen zien welke outputs correct zijn in vergelijking met de verwactte waardes en de total loss die daar uit volgt.

```python
train_perceptron(p1, pos_2x, [1, 0, 0, 0], 40) # AND
```

De XOR kan helaas nooit een total loss van 0.0 bereiken zoals de AND. Dit is ook getest met 100 epochs en nog komt de XOR total loss niet lager dan 0.25.

```python
train_perceptron(p2, pos_2x, [0, 1, 1, 0], 10) # XOR
```

## IRIS Dataset
Dit is de derde case waarin we naar de IRIS dataset kijken en hier een perceptron op gaan toepassen.
<br>Eerst laden we de data in (zowel de data als de verwachte waardes).

```python
iris = load_iris()
print(iris)
```

```python
X = iris.data
print(X)
```

Nu we de data hebben ingeladen kijken we naar de target. De target bevat echter ook de targets van de Virginica, dit moeten we dus wegfilteren zodat we alleen de Setosa en Versicolor overhouden.

```python
y = iris.target
print(y)

y = y[y != 2] # We only use Setosa and Versicolor.
print(y)

X = X[:len(y)] # Make data and target arrays equal in length.
```

## Random parameters
Zoals in de opdracht aangegeven genereren we hier willekeurige parameters voor de IRIS perceptron.

```python
random.seed(1707815)

weights = [random.random(), random.random(), random.random(), random.random()]
bias = random.random()
lr = random.random()

print(f"Weights: {weights}")
print(f"Bias: {bias}")
print(f"Learning Rate: {lr}")
```

## IRIS Perceptron en training - IRIS I
Hier maken we een perceptron met de ingevoerde willekeurige waardes. Vervolgens trainen we deze perceptron met de X data en y targets. Ik had voor 10 epochs gekozen maar hij bleek niet meer dan 4 nodig te hebben om een total loss van 0 te bereiken. Aan het einde van het trainen zien we de ge-update parameters die de perceptron nu heeft.

```python
p3 = Perceptron("IRIS", weights, bias, lr)
train_perceptron(p3, X, y, 10)
```

## IRIS Perceptron en training - IRIS II
Deze testcase is echter lastiger. De target array maakt namelijk gebruik van 0, 1 en 2. En omdat we alleen een 0 of 1 krijgen is het lastig om de Verginica te trainen (dit wordt aangegeven als 2 namelijk). Om dit te omzeilen kan ik de nul waardes weghalen, de 1 veranderen in de 0 en de 2 naar een 1. Zo kunnen we ook de X array aanpassen zodat we de records die niet Versicolor of Verginica zijn weghalen.

```python
X = iris.data
y = iris.target

count_y = np.bincount(y)

print(y)
y = y[y != 0] # We only use Versicolor and Verginica.
y[y == 1] = 0
y[y == 2] = 1
print(y)

X = X[count_y[0]:] # Make data and target arrays equal in length.
```

Nu de data klaar is kunnen we de perceptron redefinen en opnieuw trainen met de nieuwe data. Ik had voor 10 epochs gekozen maar hij bleek niet te veranderen na de eerste epoch dus heb ik hem op 3 epochs gelaten. Aan het einde van het trainen zien we de ge-update parameters die de perceptron nu heeft.

```python
p3 = Perceptron("IRIS", weights, bias, lr)
train_perceptron(p3, X, y, 3)
```
